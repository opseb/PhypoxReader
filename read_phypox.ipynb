```python
def read_phyphox_file(file_path):
    try:
        tree = ET.parse(file_path)
        root = tree.getroot()
        return root
    except ET.ParseError as e:
        print(f"Error parsing XML file: {e}")
        return None
```

```python
def extract_data_containers(root):
    data_containers = []
    data_acc = {}
    
    start = root.find(".//events").find(".//start").attrib.get("systemTime")
    pause = root.find(".//events").find(".//pause").attrib.get("systemTime")
    
    time_difference = int(pause) - int(start)
    
    print(f'START time {start}')
    print(f'PAUSE time {pause}')
    print(f'TIME difference : {time_difference}')

    for container in root.findall('.//container'):
        # print(container.attrib.items())
        items = container.attrib

        values = items.get('init').split(",")

        # print(items.get('init').split(","))
        # print(f'item keys is {items.keys()}')

        data_containers.append(container.text.strip())
        # data_containers.append(container.attrib)

        if container.text.strip() == "accY" or container.text.strip() == "accX" or container.text.strip() == "accZ" or container.text.strip() == "acc_time":
            data_acc[container.text.strip()] = values

    return data_containers, data_acc, start
```

```python
# List all .phyphox files in the subdirectories
csv_files = []

df_temp = None

for folder in folders:
    # Get .phyphox files in each folder
    files = [f for f in folder.glob('*.csv')]  # You can use `folder.glob('*.phyphox')` to search for .phyphox files
    csv_files.extend(files)


for file in csv_files:

    if file.name == "Kavenge S3 B.csv":
        df = pd.read_csv(file)
        df['t'] = pd.to_datetime(df.time, unit='ms')
        print(df.head())
        df_temp = df
        plt.rcParams["font.family"] = 'Times New Roman'
        plt.rcParams["font.size"] = 12
        plt.rcParams['xtick.minor.visible'] = True
        plt.rcParams['ytick.minor.visible'] = True
        plt.rcParams['axes.grid'] = True

        plt.figure(figsize=(12,6))
        plt.plot(df.t, df.ax, linewidth=0.75, label="ax")
        plt.plot(df.t, df.ay, linewidth=0.75, label="ay")
        plt.plot(df.t, df.az, linewidth=0.75, label="az")

        plt.legend()
        plt.show()
            
```

```python
from scipy.integrate import cumtrapz
import scipy.signal as sngl
# df = df_temp.query(" '2023-07-05 05:47:39.700' < t <'2023-07-05 05:47:44.900'")
df = df_temp.query(" '2023-07-05 05:47:47.000' < t <'2023-07-05 05:47:58.000'")

# df.az = sngl.detrend(df.az.copy().tolist())

acc = (df.az * 1000).copy().tolist()
acc = sngl.detrend(acc)
acc_original = acc / 1000
time = (df.time / 1000 ).tolist().copy()

v = cumtrapz(acc ,time)
u = cumtrapz(v,time[:-1])

plt.rcParams['lines.linewidth'] = 1

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(df.t, acc_original, label='az')
ax.set_ylabel('acc (m/s2)')
ax.legend(loc='lower left')

ax2 = ax.twinx()
ax2.plot(df.t[:-2], u, color='r', label=r'$\delta $')
ax2.set_ylabel(r'$\delta$ (mm)')
ax2.legend(loc='upper right')

plt.grid(False)
plt.show()
```
